# New plot

```python
death_dict = {}

it_id = gc_utils.iteration_name(it)

src_dat = proc_data[it_id]["source"]
ana_msk = src_dat["analyse_flag"][()] == 1

gc_src = src_dat["gc_id"][ana_msk]
grp_src = np.abs(src_dat["group_id"][ana_msk])
t_dis_src = src_dat["t_dis"][ana_msk]
t_for_src = src_dat["form_time"][ana_msk]
m_for_src = src_dat["logm_tform"][ana_msk]

death_dict["base"] = {}
death_dict["base"]["gc_id"] = gc_src
death_dict["base"]["group_id"] = grp_src
death_dict["base"]["t_dis"] = t_dis_src
death_dict["base"]["t_for"] = t_for_src
death_dict["base"]["m_for"] = m_for_src
death_dict["base"]["situ"] = np.array([0 if grp == 0 else 1 for grp in grp_src])

for i, snap in enumerate(snap_lst):
    # current snap data
    snap_id = gc_utils.snapshot_name(snap)
    snp_dat = proc_data[it_id]["snapshots"][snap_id]
    acc_msk = snp_dat["now_accreted"][()] == 1

    cur_gc_snp = snp_dat["gc_id"][acc_msk]
    cur_r_snp = snp_dat["r"][acc_msk]
    cur_t_snp = snap_pub_data[snap_pub_data["index"] == snap]["time_Gyr"].values[0]
    cur_m_snp = snp_dat["mass"][acc_msk]

    # past snap data
    if i == 0:
        gc_snp_pst = np.array([])
        r_snp_pst = np.array([])

    else:
        pst_snap_id = gc_utils.snapshot_name(np.array(snap_lst)[i - 1])
        pst_snp_dat = proc_data[it_id]["snapshots"][pst_snap_id]
        pst_acc_msk = pst_snp_dat["now_accreted"][()] == 1

        gc_snp_pst = pst_snp_dat["gc_id"][pst_acc_msk]
        r_snp_pst = pst_snp_dat["r"][pst_acc_msk]

    surv_flag = []
    r_snp = []
    t_age_snp = []
    cur_mass_snp = []

    snap_dict = {}
    for (j, gc), t_dis, t_for, m_for in zip(enumerate(gc_src), t_dis_src, t_for_src, m_for_src):
        # check if gc alive and accreted at this snapshot
        if gc in cur_gc_snp:
            surv = 1
            cur_r = cur_r_snp[cur_gc_snp == gc][0]
            cur_age = cur_t_snp - t_for
            cur_mass = cur_m_snp[cur_gc_snp == gc][0]

        else:
            if t_for > cur_t_snp:
                surv = -1
                cur_r = -1
                cur_age = -1
                cur_mass = -1

            if (t_dis < cur_t_snp) and (t_dis != -1):
                if i == 0:
                    # died before first snap analysed
                    # need to be ignored
                    surv = -2
                    cur_r = -2
                    cur_age = -2
                    cur_mass = -2

                else:
                    surv = 0
                    cur_r = death_dict[pst_snap_id]["r"][j]
                    cur_age = t_dis
                    cur_mass = 0

            # alive but not yet accreted
            else:
                surv = -1
                cur_r = -1
                cur_age = cur_t_snp - t_for
                cur_mass = -1

        surv_flag.append(surv)
        r_snp.append(cur_r)
        t_age_snp.append(cur_age)
        cur_mass_snp.append(cur_mass)

    death_dict[snap_id] = {}
    death_dict[snap_id]["surv_flag"] = np.array(surv_flag)
    death_dict[snap_id]["r"] = np.array(r_snp)
    death_dict[snap_id]["age"] = np.array(t_age_snp)
    death_dict[snap_id]["mass"] = np.array(cur_mass_snp)
```

```python
def truncate_colormap(cmap, minval=0.2, maxval=1.0, n=256):
    new_cmap = LinearSegmentedColormap.from_list(
        f"trunc({cmap.name},{minval:.2f},{maxval:.2f})", cmap(np.linspace(minval, maxval, n))
    )
    return new_cmap
```

```python
metadata = dict(title="IOM_hist_plot", artist="Finn")
writer = PillowWriter(fps=2, metadata=metadata)

fig = plt.figure(figsize=(16, 8))

gif_loc = sim_dir + sim + "/gifs/"
if not os.path.exists(gif_loc):
    os.makedirs(gif_loc)

gif_file = gif_loc + it_id + "_r_mass_form_age.gif"
with writer.saving(fig, gif_file, 100):
    for snap in snap_lst:
        snap_id = gc_utils.snapshot_name(snap)

        fig_shape = (1, 2)
        ax0 = plt.subplot2grid(shape=fig_shape, loc=(0, 0))
        ax1 = plt.subplot2grid(shape=fig_shape, loc=(0, 1))

        plt.subplots_adjust(wspace=0.2, hspace=0)

        for situ_h in [0, 1]:
            alive_msk = death_dict[snap_id]["surv_flag"] >= 0
            situ_msk = death_dict["base"]["situ"] == situ_h
            ful_msk = alive_msk & situ_msk

            age = death_dict[snap_id]["age"][ful_msk]
            sur = death_dict[snap_id]["surv_flag"][ful_msk]
            mass = 10 ** death_dict["base"]["m_for"][ful_msk]
            rad = death_dict[snap_id]["r"][ful_msk]

            c0 = []
            for a, s in zip(age, sur):
                if s == 0:
                    c0.append(a)

            plt_msk = sur == 0

            x0 = mass[plt_msk]
            y0 = rad[plt_msk]

            x1 = mass[~plt_msk]
            y1 = rad[~plt_msk]

            if situ_h == 0:
                ax = ax0
                # cmap = "Reds"
                cmap = truncate_colormap(plt.get_cmap("Reds"), 0.1, 1.0)

            else:
                ax = ax1
                # cmap = "Blues"
                cmap = truncate_colormap(plt.get_cmap("Blues"), 0.1, 1.0)

            sc = ax.scatter(x0, y0, c=c0, s=5, cmap=cmap, vmin=0, vmax=14)
            ax.scatter(x1, y1, c="k", s=5)

            ax.set_xlabel(r"GC Mass at Formation [M$_{\odot}$]")
            ax.set_ylabel("r [kpc]")

            ax.set_xscale("log")
            ax.set_yscale("log")

            ax.set_xlim([10**3.8, 10**6.7])
            ax.set_ylim([10**-2, 10**3])

            fig.colorbar(sc, ax=ax, label="Age at Disruption (Gyr)")

            ax0.text(
                0.05,
                0.95,
                "in-situ",
                color="red",
                transform=ax0.transAxes,
                ha="left",
                va="top",
                fontsize=12,
            )

            ax1.text(
                0.05,
                0.95,
                "ex-situ",
                color="blue",
                transform=ax1.transAxes,
                ha="left",
                va="top",
                fontsize=12,
            )

            time = np.round(snap_pub_data[snap_pub_data["index"] == snap]["time_Gyr"].values[0], 3)
            time_stamp = "Time: " + str(time) + " Gyr"
            ax0.text(
                0.65,
                0.95,
                time_stamp,
                color="black",
                transform=ax0.transAxes,
                ha="left",
                va="top",
                fontsize=12,
            )

        writer.grab_frame()
        fig.clear()
```
